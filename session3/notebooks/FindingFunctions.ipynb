```python
import utils
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
behavioral_path = '../data/steinmetz/raw/subject-Lederberg/session-20171209/behavioral/'
```

```python
licks = utils.merge_npy_data(behavioral_path+"licks/")

fig, ax = plt.subplots(figsize=(4, 3), dpi=150)
ax.plot(licks.sum(axis=1), c="k")
ax.set_xlabel("trials")
ax.set_ylabel("# licks")
sns.despine(trim=True)
```

```python
face = utils.merge_npy_data(behavioral_path+"face/")

fig, ax = plt.subplots(figsize=(3, 3), dpi=150)
ax.imshow(face, cmap="inferno", aspect="auto", origin="lower")
ax.set(xlabel="time", ylabel="trials")
```

```python
ws = utils.merge_npy_data(behavioral_path+"wheel_speed/")
fig, ax = plt.subplots(figsize=(4, 3), dpi=150)
ax.plot(ws.T, c="k", alpha=.1)
ax.set(xlabel="time", ylabel="wheel speed")
```

```python
import numpy as np
pupil_path = behavioral_path+'pupil/'
pupil_area_data = utils.merge_mat_data(pupil_path)

pupil_x = utils.merge_mat_data(pupil_path,'pupil_x')
pupil_y = utils.merge_mat_data(pupil_path,'pupil_y')

n_trials, n_timebins = pupil_y.shape

fig, ax = plt.subplots(figsize=(4, 3), dpi=150)
trial_idx = 29
sc = ax.scatter(pupil_x[trial_idx], pupil_y[trial_idx], c=np.linspace(0, 2.5, n_timebins), ec='k', lw=.5)
ax.set(xlabel="pupil x", ylabel="pupil y", title=f"Trial {trial_idx+1}")

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('time (s)') 
```
